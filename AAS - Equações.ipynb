```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
import math 

# Estomago
E = [500]
# Sangue
S = [0]
# Constante de absorcao do estomago para o sangue
Ks = 0.1
# Constante de eliminacao do sangue para o sistema excretor
Ke = 0.05
# âˆ†Tempo
delta_t=1e-3
# Lista tempo (horas) 
tempo = np.arange(0,24, delta_t)
```

```python
for i in range(1,len(tempo)):
    dEdt = (-1*E[i-1]*Ks)*delta_t
    dSdt = (E[i-1]*Ks - S[i-1]*Ke)*delta_t
    E.append(E[i-1] + dEdt)
    S.append(S[i-1] + dSdt)

    
plt.plot(tempo, E)
plt.plot(tempo, S)
plt.xlabel("Tempo")
plt.ylabel("ConcentraÃ§Ã£o")
plt.grid(True)
plt.show()
```
